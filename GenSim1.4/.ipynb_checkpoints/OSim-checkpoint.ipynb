This has to be changed to build a static library, remove the main program from "SORUCES", SimPop_GUs.h from "HEADERS", and to include the eigen and boost/random header files as follows:

##### pro file with g++

```python
%%file OSim.pro
######################################################################
# Automatically generated by qmake (2.01a) Tue Jul 8 16:28:42 2014
######################################################################

TEMPLATE = lib
CONFIG   = staticlib
TARGET = 
DEPENDPATH += .
INCLUDEPATH += . \
               /Users/erxingfangshui/Dropbox/CODA/eigen3 \
               /Users/erxingfangshui/Dropbox/CODA/boost

HEADERS += animal_class.h \
           cohort.h \
           ComputaionTools.h \
           genome_info.h \
           global.h \
           ped.h \
           SimPop.h
SOURCES += animal_class.cpp \
           cohort.cpp \
           ComputaionTools.cpp \
           genome_info.cpp \
           main.cpp \
           SimPop.cpp
```

Now, we generate a makefile with the following command. I found by trial and error that "-spec darwin-g++" produces a make file that works. Other options such as "-spec macx-llvm" gave an error that the "random" header ~~from boost~~ could not be found. 
* This problem with c++11 standard library is solve. The correct .pro file is shown below. Thus, we can also use llvm now.

```python
%%bash
qmake -spec darwin-g++ OSim.pro
```

```python
%ls
```

The library is now built as follows:

```python
%%bash
make
```

The library is show below as "libOSIM.a"

```python
%ls
```

```python
%%file OSim_llvm.pro
######################################################################
# Automatically generated by qmake (2.01a) Tue Jul 8 16:28:42 2014
######################################################################

TEMPLATE = lib
CONFIG   = staticlib
TARGET = 
DEPENDPATH += .
INCLUDEPATH += . \
               /Users/rohan/AeroFS/eigen3 \
               /Users/rohan/AeroFS/boost
QMAKE_CXXFLAGS += -std=c++11 -stdlib=libc++ -mmacosx-version-min=10.7 
        
# Input
HEADERS += animal_class.h \
           cohort.h \
           ComputaionTools.h \
           genome_info.h \
           ped.h \
           SimPop.h 
SOURCES += animal_class.cpp \
           cohort.cpp \
           ComputaionTools.cpp \
           genome_info.cpp 
```
