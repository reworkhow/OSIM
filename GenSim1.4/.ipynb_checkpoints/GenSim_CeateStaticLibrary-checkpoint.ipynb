# GenSim Simulation Program 

The qmake program can be used to generate a generic project file as follows.

```python
%%bash 
cat -n GenSim.pro
```

This GenSim.pro file can be used to create a Makefile to build a static library.  Correct path for Eigen and Boost should be added to INCLUDEPATH, because there libraries are used in GenSim. 

Now, we generate a makefile with the following command. I found by trial and error that "-spec darwin-g++" produces a make file that works.

```python
%%bash
#/anaconda/bin/qmake -spec darwin-g++ OSim.pro
qmake -spec /anaconda/mkspecs/darwin-g++ GenSim.pro
```

The library is now built as follows:

```python
%%bash
make
```

The library is show below as "libGenSIM.a"
