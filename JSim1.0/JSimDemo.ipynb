```julia
include ("../COSim.jl")
using COSim
```

```julia
chrLength = 1.0
numChr    = 2
numLoci   = 5
mutRate   = 0.0
locusInt  = chrLength/numLoci
mapPos   = [0:locusInt:(chrLength-0.0001)]
geneFreq = fill(0.5,numLoci)

COSim.init(numChr,numLoci,chrLength,geneFreq,mapPos,mutRate) # init COSim
```

```julia
jsim1 = COSim.startPop() # init jsim1
jsim2 = COSim.startPop() # init jsim2
nothing
```

```julia
ngen    = 3
popSize = 10
jsim1.popSample(ngen,popSize)
```

```julia
ngen    = 3
popSize = 10
jsim2.popSample(ngen,popSize)
```

```julia
jsim3 = jsim1.popNew(10) # start jsim3 by sampling from jsim1
nothing
```

```julia
jsim4 = jsim1.popNew(10) # start jsim4 by sampling from jsim1
nothing
```

```julia
jsim5 = COSim.popCross(30,jsim1,jsim2) # start jsim5 by crossing jsim1 and jsim2
nothing
```

```julia
M = jsim5.getGenotypes()
```
